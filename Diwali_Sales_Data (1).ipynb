Deleting or Removing the unnecessary columns

#Exploratory Data Analysis

```python
df.describe()
```

```python
# Sum of Amount by Zone
state_sales = df.groupby('Zone')['Amount'].sum().reset_index()


# Plotting the state-wise sales
fig = px.bar(state_sales, x='Zone', y='Amount',
             title='Total Sales by Zone',
             labels={'Amount': 'Total Sales Amount'},
             color='Amount',
             color_continuous_scale='Viridis')

fig.show()
# Most sales are generated from Central Zone such that its sales are more than combined sales of Northern, Eastern, Western Zone.
```

```python
df.groupby('Zone')['State'].unique()
```

```python
# Sum of Amount by State
state_sales = df.groupby('State')['Amount'].sum().reset_index()


# Plotting the state-wise sales
fig = px.bar(state_sales, x='State', y='Amount',
             title='Total Sales by State',
             labels={'Amount': 'Total Sales Amount'},
             color='Amount',
             color_continuous_scale='Viridis')

fig.show()
```

```python
# Sum of Amount by Gender
gender_sales = df.groupby('Gender')['Amount'].sum().reset_index()

# Plotting the gender-wise sales
fig = px.pie(gender_sales, names='Gender', values='Amount',
             title='Sales Amount by Gender',
             color_discrete_sequence=px.colors.sequential.RdBu)

fig.show()
```

```python
df['Occupation'].unique()
```

```python
women_sales_by_state = df[df['Gender'] == 'F']
# Group by Occupation and sum the Amount
women_sales_by_state = women_sales_by_state.groupby('State')['Amount'].sum().reset_index()

# Display the result
display(women_sales_by_state)

# Plotting the sales made by women by occupation
fig = px.bar(women_sales_by_state, x='State', y='Amount',
             title='Sales Made by Women by State',
             labels={'Amount': 'Total Sales Amount'},
             color='Amount',
             color_continuous_scale='Blues')

fig.show()
```

```python
men_sales_by_state = df[df['Gender'] == 'M']
# Group by Occupation and sum the Amount
men_sales_by_state = men_sales_by_state.groupby('State')['Amount'].sum().reset_index()

# Display the result
print(men_sales_by_state)

# Plotting the sales made by men by occupation
fig = px.bar(men_sales_by_state, x='State', y='Amount',
             title='Sales Made by men by state',
             labels={'Amount': 'Total Sales Amount'},
             color='Amount',
             color_continuous_scale='Blues')

fig.show()
```

```python
sales_by_gender_occupation=df.groupby(['Gender', 'Occupation'])['Amount'].sum().reset_index()
fig = px.bar(sales_by_gender_occupation,
             x='Occupation',
             y='Amount',
             color='Gender',
             title='Sales Made by Women and Men by Occupation',
             labels={'Amount': 'Total Sales Amount'},
             barmode='group',  # Grouped bar chart for comparison
             color_discrete_map={'F': 'pink', 'M': 'blue'})  # Custom colors for Gender

fig.show()
```

```python
# Sum of Amount by Age Group
age_group_sales = df.groupby('Age Group')['Amount'].sum().reset_index()

# Plotting the age group-wise sales
fig = px.bar(age_group_sales, x='Age Group', y='Amount',
             title='Sales Amount by Age Group',
             labels={'Amount': 'Total Sales Amount'},
             color='Amount',
             color_continuous_scale='Plasma')

fig.show()
```

```python
age_group_26_35 = df[df['Age Group'] == '26-35']

# Group by Occupation and count the number of people in each sector
sectors_by_age_group = age_group_26_35['Occupation'].value_counts().reset_index()
sectors_by_age_group.columns = ['Occupation', 'Count']

# Display the result
print(sectors_by_age_group)

# Plotting the number of people aged 26-35 by occupation
fig = px.bar(sectors_by_age_group, x='Occupation', y='Count',
             title='Sectors for Age Group 26-35',
             labels={'Count': 'Number of People'},
             color='Count',
             color_continuous_scale='Viridis')

fig.show()
```

```python
sales_by_product_category = df.groupby('Product_Category')['Amount'].sum().reset_index()

# Display the result
print(sales_by_product_category)

# Plotting the sales by product category using a pie chart
fig = px.pie(sales_by_product_category, names='Product_Category', values='Amount',
             title='Sales Distribution by Product Category',
             color='Amount',
             )

fig.show()
```

```python
women_df = df[df['Gender'] == 'F']

# Group by Marital_Status and sum the Amount
sales_by_marital_status = women_df.groupby('Marital_Status')['Amount'].sum().reset_index()

# Create the bar plot
fig = px.bar(sales_by_marital_status, x='Marital_Status', y='Amount',
             title='Sales by Marital Status for Women',
             labels={'Marital_Status': 'Marital Status', 'Amount': 'Total Sales Amount'},
             color='Marital_Status',
             color_discrete_map={0: 'blue', 1: 'pink'})

# Show the plot
fig.show()
```

```python
sales_by_gender_marital_status_orders = df.groupby(['Gender', 'Marital_Status'])['Orders'].sum().reset_index()

fig = px.bar(sales_by_gender_marital_status_orders,
             x='Marital_Status',
             y='Orders',
             color='Gender',
             title='Total Number of Orders by Gender and Marital Status',
             labels={'Orders': 'Total Number of Orders', 'Marital_Status': 'Marital Status'},
             barmode='group',
             color_discrete_map={'F': 'red', 'M': 'blue'})

fig.show()
```

```python
# Filter the dataset for married women (Marital_Status = 1) and Gender = 'F'
married_women_spending = df[(df['Marital_Status'] == 1) & (df['Gender'] == 'F')]

# Group by Product Category and sum the Amount
spending_by_product_category = married_women_spending.groupby('Product_Category')['Amount'].sum().reset_index()

# Display the result
print(spending_by_product_category)

# Plotting the spending by married women on product categories using a pie chart
fig = px.pie(spending_by_product_category, names='Product_Category', values='Amount',
             title='Spending by Married Women on Product Categories')

fig.show()
```

```python
# Filter the dataset for married men (Marital_Status = 1) and Gender = 'M'
married_men_spending = df[(df['Marital_Status'] == 1) & (df['Gender'] == 'M')]

# Group by Product Category and sum the Amount
spending_by_product_category = married_men_spending.groupby('Product_Category')['Amount'].sum().reset_index()

# Display the result
print(spending_by_product_category)

# Plotting the spending by married men on product categories using a pie chart
fig = px.pie(spending_by_product_category, names='Product_Category', values='Amount',
             title='Spending by Married men on Product Categories')

fig.show()
```

```python
fig = px.scatter(df, x='Age', y='Amount', color='Gender',
                 title='Amount Spent vs. Age by Gender',
                 labels={'Amount': 'Total Sales Amount', 'Age': 'Customer Age'},
                 trendline='ols')  # Optional: Add a trendline
fig.show()
```

```python
fig = px.scatter(df, x='Orders', y='Amount', color='Product_Category',
                 title='Orders vs. Amount Spent by Product Category',
                 labels={'Orders': 'Number of Orders', 'Amount': 'Total Sales Amount'})
fig.show()

```

```python
df['Product_Category'].unique()
```

```python
# Group by Age Group and Product Category, then count the number of orders
interest_by_age_category = df.groupby(['Age Group', 'Product_Category'])['Orders'].sum().reset_index(name='Count')

# Create a treemap to visualize the interest
fig = px.treemap(interest_by_age_category,
                 path=['Age Group', 'Product_Category'],
                 values='Count',
                 title='Number of Orders by Age Group and Product Category',
                 )
fig.show()
```

#Conclusions/Insights Drawn



*   Most sales are generated from Central Zone such that its sales are more than combined sales of Northern, Eastern, Western Zone.
*   Top states that generated Highest number of sales are Uttarpradesh, Maharashtra, Karnataka, Delhi and Madhyapradesh Consequently.
*   70% of the sales are generated from women that means women are showcasing high purchsing power and willing to buy and these women are mostly from uttarpradesh, Maharashtra and Karnataka.
*Men and Women who are working in sectors like IT, Heathcare, Aviation and Banking are contributing Mostly to the sales. Mostly Women are Highly dominating the charts.
* The Age Group of 18-45 should be Highly Targeted Group and Age group of 26-35 are showing high purchasing power compared to remaining groups and are mostly working in IT, Heathcare, Aviation and Banking Sectors.     
* In all the 18 Categories of products sold Food, Clothing & Apparel,
 Electronics & Gadgets,Footwear & Shoes are contributing around 75% of the sales.
* we could observe single orders and multiple orders are contributing relatively equally to sales.
* Married women are showcasing High purchasing power and Spending more than the whole men. They belongs to IT, Heathcare, Aviation and Banking Sectors which indicating their High income levels and mostly from uttarpradesh, Maharashtra and Karnataka.
